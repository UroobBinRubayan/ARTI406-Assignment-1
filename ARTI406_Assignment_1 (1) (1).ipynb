# Assignment 1: Exploratory Data Analysis (EDA)

EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

### Descriptive summary Statistics

In this step, the dataset was explored to understand its structure and contents.

This step helps to gain an initial understanding of the dataset before performing data cleaning and analysis.

### Univariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df['Test_Score'], bins=20)
plt.title("Distribution of Test Scores")
plt.show()
```

- This histogram shows the distribution of student test scores. Most students have scores in the mid-range, indicating average academic performance. There are fewer students with very low or very high scores.

## Bivariate Analysis

### Study Hours vs Test Score Analysis

```python
sns.scatterplot(x='Study_Hours', y='Test_Score', hue='Gender', data=df)
plt.title("Study Hours vs Test Score by Gender")
plt.show()
```

Average Test Score by Gender

```python
plt.figure(figsize=(8,5))
df.groupby('Gender')['Test_Score'].mean().plot(kind='bar')
plt.title("Average Test Score by Gender")
plt.ylabel("Test Score")
plt.show()
```

### Top Students Based on Test Scores

```python
top_students = df.sort_values(by='Test_Score', ascending=False).head(10)

top_students[['Student_ID', 'Test_Score']]
```

This table shows the top students based on their test scores. It helps identify high-performing students in the dataset.

### Correlation Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.heatmap(df[['Study_Hours', 'Attendance(%)', 'Test_Score']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

- Correlation close to 1 = strong positive relationship
- Close to 0 = weak relationship

## Time-Based Analysis
### Relationship Between Study Hours and Average Test Scores

```python
avg_score = df.groupby('Study_Hours')['Test_Score'].mean()

plt.figure(figsize=(10,5))
avg_score.plot()
plt.title("Average Test Score by Study Hours")
plt.ylabel("Test Score")
plt.xlabel("Study Hours")
plt.show()
```
